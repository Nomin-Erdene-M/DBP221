# Ирц

```python
import pandas as pd
import os
os.chdir(r"F:\UFE\4th semester\DBP221\Biy daalt")
irts=pd.read_excel('Бие даалт - 1.xlsx').dropna(axis=1, how='all')
irts['Нийт -10']=(irts[irts == 'и'].count(axis=1)*10/32)
irts
```

# Семинар

```python
seminar=pd.read_excel('Бие даалт - 1.xlsx', sheet_name=1).dropna(axis=1, how='all').fillna(0)
seminar['Нийт -25'] = (seminar.iloc[:,3:22].sum(axis=1))*25/19
seminar
```

# Шалгалт

```python
exam=pd.read_excel('Бие даалт - 1.xlsx', sheet_name=2).dropna(axis=1, how='all').fillna(0)
exam['Нийт']=exam['Шалгалт - 1']+exam['Шалгалт - 2']+exam['Шалгалт - 3']
exam
```

# Бие даалт

```python
biydaalt=pd.read_excel('Бие даалт - 1.xlsx', sheet_name=3).dropna(axis=1, how='all').fillna(0)
biydaalt['Нийт'] = biydaalt['Бие даалт - 1']+biydaalt['Бие даалт - 2']
biydaalt
```

# Хүмүүжил хандлага

```python
attitude=pd.read_excel('Бие даалт - 1.xlsx', sheet_name=4).dropna(axis=1, how='all').fillna(0)
attitude['Нийт'] = attitude.iloc[:,3:7].sum(axis=1)
attitude
```

# Нэмэлт

```python
nemelt=pd.read_excel('Бие даалт - 1.xlsx', sheet_name=5).dropna(axis=1, how='all').fillna(0)
nemelt['Нийт']=nemelt['Нэмэлт оноо']
nemelt
```

# Нийт дүн

```python
niit=pd.read_excel('Бие даалт - 1.xlsx', sheet_name=6).dropna(axis=0, how='all').dropna(axis=1, how= 'all')
niit['Ирц-10']=irts['Нийт -10']
niit['Семинар-25']=seminar['Нийт -25']
niit['Шалгалт-30']=exam['Нийт']
niit['Бие даалт-25']=biydaalt['Нийт']
niit['ХХ-10']=attitude['Нийт']
niit['Нэмэлт оноо']=nemelt['Нийт']

niit['Нийт']=niit.iloc[:,3:9].sum(axis=1)
niit.rename(columns={'Unnamed: 11':'Үнэлгээ'},inplace=True)
niit.at[28,'Нийт'] = niit['Нийт'].mean()
niit['Үнэлгээ'] = niit['Нийт'].map(lambda x: 'F' if 0 <= x < 59.5 else 
                                      'D-' if 59.5 <= x < 62.5 else 
                                      'D' if 62.5 <= x < 66.5 else 
                                      'D+' if 66.5 <= x < 69.5 else 
                                      'C-' if 69.5 <= x < 72.5 else
                                      'C' if 72.5 <= x < 76.5 else
                                      'C+' if 76.5 <= x < 79.5 else
                                      'B-' if 79.5 <= x < 82.5 else
                                      'B' if 82.5 <= x < 86.5 else
                                      'B+' if 86.5 <= x < 89.5 else
                                      'A-' if 89.5 <= x < 95.5 else
                                      'A' if 95.5 <= x <= 100 else None)
data1=niit['Үнэлгээ'][:28]
a=data1.value_counts()['A']+data1.value_counts()['A-']
b=data1.value_counts()['B']+data1.value_counts()['B+']+data1.value_counts()['B-']
c=data1.value_counts()['C']+data1.value_counts()['C+']
d=data1.value_counts()['D+']
f=data1.value_counts()['F']
niit.rename(columns={'Unnamed: 13':'Оюутны тоо'},inplace=True)
niit.at[0,'Оюутны тоо'] = a
niit.at[1,'Оюутны тоо'] = b
niit.at[2,'Оюутны тоо'] = c
niit.at[3,'Оюутны тоо'] = d
niit.at[4,'Оюутны тоо'] = f
niit=niit.fillna(' ')
niit
```

# Чанар, амжилт, хамгийн их, хамгийн бага, дундаж

```python
irts1 = pd.Series({'Хамгийн өндөр дүн': irts['Нийт -10'].max(), 'Хамгийн бага дүн':irts['Нийт -10'].min(), 'Дундаж':irts['Нийт -10'].mean()})
seminar1 = pd.Series({'Хамгийн өндөр дүн': seminar['Нийт -25'].max(), 'Хамгийн бага дүн':seminar['Нийт -25'].min(), 'Дундаж':seminar['Нийт -25'].mean()})
exam1 = pd.Series({'Хамгийн өндөр дүн': exam['Нийт'].max(), 'Хамгийн бага дүн':exam['Нийт'].min(), 'Дундаж':exam['Нийт'].mean()})
biydaalt1 = pd.Series({'Хамгийн өндөр дүн': biydaalt['Нийт'].max(), 'Хамгийн бага дүн':biydaalt['Нийт'].min(), 'Дундаж':biydaalt['Нийт'].mean()})
attitude1 = pd.Series({'Хамгийн өндөр дүн': attitude['Нийт'].max(), 'Хамгийн бага дүн':attitude['Нийт'].min(), 'Дундаж':attitude['Нийт'].mean()})
nemelt1 = pd.Series({'Хамгийн өндөр дүн': nemelt['Нийт'].max(), 'Хамгийн бага дүн':nemelt['Нийт'].min(), 'Дундаж':nemelt['Нийт'].mean()})
niit1 = pd.Series({'Хамгийн өндөр дүн': niit['Нийт'].max(), 'Хамгийн бага дүн':niit['Нийт'].min(), 'Дундаж':niit['Нийт'].mean()})
newSheetScore = pd.DataFrame({'Ирц':irts1,'Семинар': seminar1, 'Шалгалт':exam1,'Бие даалт': biydaalt1, 'Хүмүүжил хандлага':attitude1, 'Нэмэлт оноо': nemelt1, 'Нийт дүн':niit1})

success=(1-niit['Оюутны тоо'][4]/niit['Оюутны тоо'][5])*100
chanar=((niit['Оюутны тоо'][0]+niit['Оюутны тоо'][1])/niit['Оюутны тоо'][5])*100
data2= pd.Series({'Хичээлийн амжилт': success , 'Хичээлийн чанар': chanar })

data3=pd.concat([data2,newSheetScore.T])
data3=data3.fillna(' ')
data3
```

# Excel

```python
writer = pd.ExcelWriter('Biydaalt_M.Nomin-Erdene.xlsx', engine = 'xlsxwriter')
irts.to_excel(writer, sheet_name='Ирц')
seminar.to_excel(writer, sheet_name='Семинар')
exam.to_excel(writer, sheet_name='Шалгалт')
biydaalt.to_excel(writer, sheet_name='Бие даалт')
attitude.to_excel(writer, sheet_name='Хүмүүжил хандлага')
nemelt.to_excel(writer, sheet_name='Нэмэлт оноо')
niit.to_excel(writer, sheet_name='Нийт')
data3.to_excel(writer, sheet_name='Нэмэлт тооцоолол')
writer.save()
```
