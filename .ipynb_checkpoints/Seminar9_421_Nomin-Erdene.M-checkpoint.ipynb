```python
import pandas as pd
name = pd.Series(['Bob', 'Jake', 'Lisa', 'Sue'])
name
```

```python
#1 len()
name.str.len()
# Элементийн уртыг олно.
```

```python
#2 lower()
name.str.lower()
# Бүх үсгийг жижиг болгоно.
```

```python
#3 translate()
data1 = "Jake"
data2 = data1.maketrans("k", "n")
print(data1.translate(data2))
# Үсгийг өөр үсгээр солино.
```

```python
#4 islower()
name.str.islower()
# Текст бүхэлдээ жижиг үсгээр бичигдсэн эсэхийг шалгана.
```

```python
#5 ljust()
a = "Lisa"
b = a.ljust(7, "a")
print(b)
# Заасан урттай зайнд тухайн үгийг зүүн тийш шахаж оруулна, хүсвэл дурын үсгээр баруун талын хоосон зайг дүүргэж болно.
```

```python
#6 upper()
name.str.upper()
# Бүх үсгийг томоор бичнэ.
```

```python
#7 startswith()
txt = "My name is Nomin."
x = txt.startswith("name")
print(x)
# Текст заасан үгээр эхэлж байгаа эсэхийг шалгана.
```

```python
#8 isupper()
name.str.isupper()
# Текст бүхэлдээ том үсгээр бичигдсэн эсэхийг шалгана.
```

```python
#9 rjust()
c = "Bob"
d = c.rjust(10,"i")
print(d)
# Заасан урттай зайнд тухайн үгийг баруун тийш шахаж оруулна, хүсвэл дурын үсгээр зүүн талын хоосон зайг дүүргэж болно.
```

```python
#10 find()
txt = "Hello, welcome to my world."
x = txt.find("lo")
print(x)
# Хайсан үг, үсгийн эхэлж байгаа индексийг тогтооно.
```

```python
#11 endswith()
txt = "My name is Nomin."
x = txt.endswith(".")
print(x)
# Текст тухайн төгсгөлөөр төгсөж байгаа эсэхийг шалгана.
```

```python
#12  isnumeric()
e = "118"
f = e.isnumeric()
print(f)
# Текст бүхэлдээ тоон утгатай эсэхийг шалгана.
```

```python
#13 center()
txt = "bob"
x = txt.center(20,"i")
print(x)
# Заасан зайнд тухайн үгийг голд нь байрлуулж, хүсвэл дурын үсгээр хоосон зайг дүүргэж болно.
```

```python
#14 rfind()
txt = "My name is Nomin."
x = txt.rfind("m")
print(x)
# Хамгийн баруун талд байгаа заасан үсэг хэд дэх индекс дээр байгааг шалгана.
```

```python
#15 get()
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = car.get("model")
print(x)
# Заасан key утгаас value утгыг олно.
```

```python
#16 isalnum()
txt = "Company12"
x = txt.isalnum()
print(x)
# Текстэнд үсэг, тоо 2уулаа байгаа эсэхийг шалгана.
```

```python
#17 isdecimal()
s = "1156"
print(s.isdecimal())
# Текст нь тоон утга эсэхийг шалгаж байна.
```

```python
#18 zfill()
txt = "50"
x = txt.zfill(10)
print(x)
# Заасан утгыг баруун талаас нэмээд заасан хэмжээнд хүртэл нь утга нөхнө.
```

```python
#19 index()
fruits = ['apple', 'banana', 'cherry']
x = fruits.index("cherry")
x
# Заасан утгын индексийг харуулна.
```

```python
#20 isalpha()
txt = "CompanyX"
x = txt.isalpha()
print(x)
# Текстэнд байгаа бүх утга үсэг эсэхийг шалгана.
```

```python
#21 split()
txt = "My name is Nomin."
x = txt.split()
print(x)
# Бүх утгыг тусад нь салгаж авна.
```

```python
#22 strip()
txt = "     name     "
x = txt.strip()
print("My", x, "is Nomin.")
# Эхлэл төгсөлд буй илүү зайг хасна
```

```python
#23 rindex()
txt = "My name is Nomin."
x = txt.rindex("y")
print(x)
# Заасан үг, үсэг хэд дэх индекст буйг заана.
```

```python
#24 isdigit()
txt = "50800"
x = txt.isdigit()
print(x)
# Бүх утга тоо эсэхийг шалгана.
```

```python
#25 rsplit()
txt = "apple, banana, cherry"
x = txt.rsplit(", ")
print(x)
# Утгуудыг бүгдийг нь тусад нь салгаж авна.
```

```python
#26 rstrip()
txt = "     name     "
x = txt.rstrip()
print("My", x, "is Nomin.")
# Зайг баруун талаас шахна.
```

```python
#27 capitalize()
name.str.capitalize()
# Бүх үгийг томоор эхлүүлж бичнэ.
```

```python
#28 isspace()
txt = "   "
x = txt.isspace()
print(x)
# Текст бүхэлдээ хоосон эсэхийг шалгана.
```

```python
#29 partition()
txt = "My name is Nomin."
x = txt.partition("is")
print(x)
# Текстийг сонгогдсон утгаас урагшаа болон хойшоо тасална.
```

```python
#30 lstrip()
txt = "     name     "
x = txt.lstrip()
print("My", x, "is Nomin.")
# Зайг зүүн талаас шахна.
```

```python
#31 swapcase()
txt = "Hello My Name Is NOMIN."
x = txt.swapcase()
print(x)
# Том үсгийг жижиг, жижиг үсгийг том болгож солино.
```

```python
#32 istitle()
txt = "Hello My Name Is Nomin."
x = txt.istitle()
print(x)
# Бүх үг томоор эхэлж байгаа (бусад үсэг нь жижиг) эсэхийг шалгана. 
```

```python
#33 rpartition()
txt = "My name is Nomin."
x = txt.rpartition("is")
print(x)
# Сонгогдсон утгаас урагшаа болон хойшоо баруун талаас тасална
```
